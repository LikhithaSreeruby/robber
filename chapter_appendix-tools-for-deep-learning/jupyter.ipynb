```python
month = int(input("Enter month : "))
year = int(input("Enter year: "))

noofdays = [31,28,31,30,31,30,31,31,30,31,30,31]

if (year%4==0 and year%100!=0) or (year%400==0):
    noofdays[1] = 29

month_odd = [0,3,3,6,1,4,6,2,5,0,3,5]

y = year - 1
odd = y + y//4 - y//100 + y//400

odd = odd + month_odd[month-1]

if month > 2 and ((year%4==0 and year%100!=0) or (year%400==0)):
    odd = odd + 1

start = odd % 7

print("\nSun Mon Tue Wed Thu Fri Sat")

for i in range(start):
    print("    ", end="")

for d in range(1, noofdays[month-1] + 1):
    print(f"{d:3}", end=" ")
    if (d + start) % 7 == 0:
        print()
```

```python
sentence = input("Enter a sentence: ")

vowels = "aeiouAEIOU"
final = ""

for ch in sentence:
    if ch not in vowels:
        final = final + ch

print("Sentence after removing vowels:", final)
```

```python
n = int(input("Enter no.of names: "))

names = []

for i in range(n):
    name = input("Enter name: ")
    names.append(name)

names.sort()

print("order:")
for name in names:
    print(name)
```
